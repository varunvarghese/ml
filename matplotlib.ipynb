```python
x = np.linspace(-1.4, 1.4, 30)
```

```python
plt.plot(x, x, 'g--', 
         x, x**2, 'r:', 
         x, x**3, 'b^')
plt.show()
```

```python
plt.plot(x, x, 'g--') 
plt.plot(x, x**2, 'r:') 
plt.plot(x, x**3, 'b^')
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1 = plt.plot(x, x, 'g--')
line2 = plt.plot(x, x**2, 'r:')
line3 = plt.plot(x, x**3, 'b^')
#line1.set_linewidth(3.0)
#line1.set_dash_capstyle("round")
#line3.set_alpha(0.2)
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1 = plt.plot(x, x, 'g--', linewidth=3.0, dash_capstyle="round")
line2 = plt.plot(x, x**2, 'r:')
line3 = plt.plot(x, x**3, 'b^', alpha=0.2)
plt.show()
```
